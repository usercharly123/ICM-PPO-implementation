```python
from tensorboard.backend.event_processing import event_accumulator
import matplotlib.pyplot as plt
```

```python
# Load TensorBoard event data
ea = event_accumulator.EventAccumulator('events/')
ea.Reload()

# Extract scalar data
events_extr = ea.Scalars('Mean_extr_reward_per_1000_steps')
events_intr = ea.Scalars('Mean_intr_reward_per_1000_steps')

# Convert to lists
steps_extr = [e.step for e in events_extr]
values_extr = [e.value for e in events_extr]
steps_intr = [e.step for e in events_intr]
values_intr = [e.value for e in events_intr]

# Create subplots
fig, axs = plt.subplots(2, 1, figsize=(10, 8), sharex=True)

# Extrinsic reward plot
axs[0].plot(steps_extr, values_extr, label='Extrinsic Reward', color='blue')
axs[0].set_ylabel('Mean Extrinsic Reward')
axs[0].set_title('Mean Extrinsic Reward per 1000 Steps')
axs[0].grid(True)

# Intrinsic reward plot
axs[1].plot(steps_intr, values_intr, label='Intrinsic Reward', color='green')
axs[1].set_xlabel('Training Steps')
axs[1].set_ylabel('Mean Intrinsic Reward')
axs[1].set_title('Mean Intrinsic Reward per 1000 Steps')
axs[1].grid(True)

plt.tight_layout()
plt.show()
```
